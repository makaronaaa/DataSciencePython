```python
br3 = data.replace('na', np.NaN)
br3['reps_lower'] = pd.to_numeric(br3['reps_lower'], errors='coerce')
br3['reps_lower'] = round(br3['reps_lower'].fillna(br3['reps_lower'].mean()).astype('float'),2)
br3
```

```python
br4 = br3.replace('na', np.NaN)
br4['reps_upper'] = pd.to_numeric(br4['reps_upper'], errors='coerce')
br4['reps_upper'] = round(br4['reps_upper'].fillna(br4['reps_upper'].mean()).astype('float'),2)
br4
```

```python
br4['obey'] = br4['obey'].replace('%','', regex=True)
br4['obey'] = pd.to_numeric(br4['obey'],errors='coerce')
br4['obey']=br4['obey'].fillna(br4['obey'].mean())
br4
```

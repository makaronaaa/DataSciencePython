```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sb
from sklearn import metrics
data=pd.read_csv('/content/sample_data/GOLD_weekly_return_volatility_labeled.csv')
data
```

```python
X_train = data[["mean_return","volatility"]][data["Year"]==2020].values
Y_train = data[["label"]][data["Year"]==2020].values

df = data[data["Year"]==2021].reset_index(drop=True)
X_test = df[["mean_return","volatility"]]
Y_test = df["label"]
```

```python
gnb = GaussianNB()
y_pred = gnb.fit(X_train, Y_train).predict(X_test)
accuracy = round(100*np.mean(y_pred == Y_test) , 2)
print("GNB Accuracy Year2: ", accuracy, "%")
```

```python
cf = confusion_matrix(Y_test , y_pred)
print("confusion matrix: considering green is + \n",  cf)   
tp = cf[0][0]
fn = cf[0][1]
fp = cf[1][0]
tn = cf[1][1]
print("TP=", tp,"    FP=",  fp,"    FN=",  fn,"   TN=",  tn)

tpr = tp / (tp + fn)
tnr = tn / (tn + fp)
acc = (tp + tn) / (tp + tn + fp + fn)
print("TPR=", tpr, "    TNR=", tnr, "    ACC=", acc)
```

```python
df["gnb"] = y_pred
df[ :4]
```

```python
# BUY & HOLD stgy
value = 100
for i in range(len(df)):
    value = round(value*(100 + df['mean_return'][i])/100, 2)
        
print("Final value of portfolio after BUY AND HOLD for year-2: $", value)    
print("annual return = ", round((value-100), 2), '%')
```

```python
#TRADING BY GNB PREDICTED LABELS for Year 2
value = 100
for i in range(len(df)):
    if df['gnb'][i] == 'green':
        value = round(value*(100 + df['mean_return'][i])/100, 2)
        
print("Final Value of Portfolio after trading  ", value)    
print("annual return = ", round((value-100), 2), '%')
```

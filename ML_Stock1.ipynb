```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#df=pd.read_csv('/content/sample_data/GOLD_weekly_return_volatility_labeled.csv')
df1=pd.read_csv('/content/sample_data/GOLD_weekly_return_detailed.csv')

###########################################Buy and hold trading strategy for year-2021

#implement a ”buy-and-hold” strategy for year 2 - you invest $100 on the first trading day (at the opening price) and sell
#at the last trading day at the (adjusted) closing price.

df2 =df1[df1["Year"]==2021].reset_index(drop=True)
invest=100
profit=((invest/df2['Open'][0])*df2['Adj Close'].iloc[-1])-invest
ROI_profit_pc = (profit/invest)*100

print(profit)
print(ROI_profit_pc)

##################################################### 
# df2 =df1[df1["Year"]==2021].reset_index(drop=True)
# invest=100
# openPR=df2['Open'][0]
# stock_purchased=invest/openPR
# redemption_repayment=df2['Adj Close'].iloc[-1]*stock_purchased
# profit=redemption_repayment-invest
# ROI_profit_pc= (profit/invest)*100
# print(profit)
# print(ROI_profit_pc)
```

```python
df=pd.read_csv('/content/sample_data/GOLD_weekly_return_volatility_labeled.csv')  

##σ: standard deviation ("volatility") of daily returns
#mean_return=((CLosingPrice_Friday-OPeningPr_Mon)/OpPrice_Monday)*100
#volatility=STDEV=(mean of returns)SQRT/Trading days-1

#implement a trading strategy based on your labels for year 2 and compare the performance with the ”buy-and-hold” strategy.
df3=df[df['Year']==2021].reset_index(drop=True)

invest=100
for i in range(len(df3)):
    if df3['label'][i] == 'green':
     invest = invest*(100 + df3['mean_return'][i])/100
        
print("Final $$ after trading by labels for year-2: ", invest)    
print("annual return = ", round((invest-100), 2), '%')
df3
```

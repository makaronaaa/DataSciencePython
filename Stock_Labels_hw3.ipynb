```python
import pandas as pd
import numpy as np
df = pd.read_csv(r"C:\Users\maka\OneDrive\Desktop\Assignment_3\weekly_return_volatility_lab.csv")
df
```

```python
import matplotlib.pyplot as plt 
x =df['mean_return'][df['Year'] == 2020]
y =df['volatility'][df['Year'] == 2020]
label =df['label'][df['Year'] == 2020]
for i in range(len(x)):
    plt.scatter(x[i],y[i],c=label[i],s=50, linewidth=0)
plt.title("Volatility vs Mean Returns for 2020")
plt.xlabel("Mean Returns")
plt.ylabel("Volatility")
```

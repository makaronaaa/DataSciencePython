```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import seaborn as sb
from sklearn import metrics

df=pd.read_csv('/content/sample_data/data_banknote_authentication.csv')
df['color'] = np.where(df['class']== 0 , 'green', 'red')
df
```

```python
X = df[["variance","skewness","curtosis", "entropy"]].values # Features
Y = df[['color']].values # Target variable
X_train,X_test,X1_train,X1_test,X2_train,X2_test,X3_train,X3_test,X4_train,X4_test,Y_train,Y_test = train_test_split(X,X1,X2,X3,X4,Y,test_size = 0.5)

#scale
scaler = StandardScaler()
scaler.fit(X_train)
xtrain = scaler.fit_transform(X_train) 
xtest = scaler.transform(X_test)

classifier = LogisticRegression()
# fit the model with data
classifier.fit(X_train, Y_train)
y_pred = classifier.predict(X_test)

from sklearn.metrics import accuracy_score
accur=accuracy_score(Y_test, y_pred)
print ("Accuracy : ", round(100*accur),2)
```

```python
#confusion metrix- accuracy btw positive and negative labels
cm = confusion_matrix(Y_test, y_pred)  #check the model performance on predicted values and actual values.
tp = cm[0][0]
fn = cm[0][1]
fp = cm[1][0]
tn = cm[1][1]
print ("Confusion Matrix : \n", cm)
print("TP=", tp,"    FP=",  fp,"    FN=",  fn,"   TN=",  tn)
tpr = tp / (tp + fn)
tnr = tn / (tn + fp)
acc = (tp + tn) / (tp + tn + fp + fn)
print("TPR=", tpr, "    TNR=", tnr, "    ACC=", acc)

X_accuracy= round(100*(tp + tn)/(tp + tn + fp + fn), 2)
print("Accuracy",X_accuracy)
```

-a statistical analysis method to predict a binary outcome, such as yes or no, based on prior observations of a data set(values)

-generally no scalling(we don't calculate distance), but to compare features we scale.
